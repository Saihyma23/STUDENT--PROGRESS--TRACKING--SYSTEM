```
print("*STUDENT PROGRESS TRACKING SYSTEM*")
print("\n")
NAME=input("STUDENT NAME:-")
CLASS=input("STUDENT CLASS:-")
ROLL=input("STUDENT ROLL NO:-")
SUBJECT=int(input("Enter your subjects:-"))
n=SUBJECT
marks=[]
for i in range(0,n):
    mark = int(input(f"enter subject{i+1} your marks:- "))
    marks.append(mark)
    if mark <= 34:
        print(f"You have failed in subject {i+1}")

print(f"marks:-{marks}")
Total=0
for i in range(0,n):
    Total=marks[i]+Total
print(f"Total:-{Total}")
Average=0
Average=Total/n
print(f"Average:-{Average}")

#print("GRADE:-")
if Average>=90 and Average<=100:
    Grade="A"
elif Average>=75 and Average<=89:
    Grade="B"
elif Average>=50 and Average<=75:
    Grade="C"
elif Average>=35 and Average<=49:
    Grade="D"
elif Average<=34:
    print(f"you are Failed{Average}")
else:
    print("you are pass")

print(f"Grade:-{Grade}")

if any(mark <= 34 for mark in marks):
    status = "FAIL"
else:
    status = "PASS"
print(f"Result: {status}")
print("\n")
print("*STUDENT PROGRESS REPORT*")
print("\n")
print(NAME)
print(CLASS)
print(ROLL)
print(Total)
print(Average)
print(Grade)
print("\n")
print("*MENU*")
print("\n")
print("what do you want to do next ")
Choice=input("enter your Choice:-")
while Choice!="3":
    if Choice=="1":
        print(f"marks:-{marks}")
        Choice=input("enter your Choice:-")
        print("\n")
    elif Choice=="2":
        print("\n")
        print("\n*STUDENT PROGRESS REPORT*")
        print(NAME)
        print(CLASS)
        print(ROLL)
        print(f"marks:-{marks}")
        print(f"Total:-{Total}")
        print(f"Average:-{Average}")
        print(f"Grade:-{Grade}")
        print(f"Result: {status}")
        Choice=input("enter your Choice:-")
        #break
        print("\n")
    elif Choice=="3":
        exit
print("THANK YOU*")
```
